# Projects in data science: GitHub

## Part 1: Setup Git and GitHub
1. [Install Git](https://docs.github.com/en/get-started/getting-started-with-git/set-up-git#setting-up-git)
2. Create your GitHub account
3. [Setup Personal Access Token (PAT)](https://www.geeksforgeeks.org/git/how-to-generate-personal-access-token-in-github/)
4. Verify your installation with a simple Hello World repository as for example shown during the lecture

## Part 2: First steps with Git
Do this part with your group


**Exercices** :
1. One member creates a new repository on GitHub with a *README.md* file and add the other members as collaborators
2. Every member can then do the following instruction **outside the browser**:
    * Get the repository from GitHub
    * Add their name to the *README.md* file
    * Create a new file
    * push the changes to the remote repository, resolving potential conflicts if needed
3. Create a gitlog.txt to see the history of your repo with the command git log > gitlog.txt


## Part 3: Work with an existing folder
You can do this part alone

**Exercises**:
1. Create a local folder
2. Create a empty repository on GitHub and link your local folder to it
3. Create config.py containing the following line, do NOT replace "YOUR EMAIL", it will be in another step.
    ```python
    email="YOUR_EMAIL"
    ```

4. Push config.py on the remote server

5. Create script.py with the following code
    ```python
    from config import email
    print(f"My email is: {email}")
    ```

6. Replace "YOUR_EMAIL" by your real email
7. Create a .gitignore file to ensure changes in config.py won't be pushed to the remote repository
8. Push what's possible to the remote repository
